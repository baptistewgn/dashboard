```python
import pandas as pd
from pathlib import Path
import datetime as dt
```

```python
data_path = Path("/workspaces/codespaces-blank/data/table.parquet") #to be updated to the correct path

def load_prices():
    if data_path.exists():
        return pd.read_parquet(data_path)
    return pd.DataFrame()

df = load_prices()
print(df.head())
```

```python
data_path = Path("/workspaces/codespaces-blank/data/sample.parquet")

df_old = pd.DataFrame([
    {"price": 100},
    {"price": 110},
    {"price": 120},
])
df_old.index = [dt.datetime(2024, 1, 1), dt.datetime(2024, 1, 2), dt.datetime(2024, 1, 3)]

df_new = pd.DataFrame([
    {"price": 115},
    {"price": 110},
    {"price": 105},
])
df_new.index = [dt.datetime(2024, 1, 3), dt.datetime(2024, 1, 4), dt.datetime(2024, 1, 5)]

df_all = pd.concat([df_old, df_new])
df_all = df_all[~df_all.index.duplicated(keep="last")]

def save_prices(df):
    df.sort_index(inplace=True)
    df.to_parquet(data_path)

save_prices(df_all)
df = load_prices()
print(df.head())
```
