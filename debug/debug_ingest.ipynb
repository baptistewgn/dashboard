```python
import yfinance as yf
import pandas as pd
# from config import tickers, start_date


"""simulate from config import tickers, start_date"""


tickers = {
    "Apple": "AAPL",
    "Microsoft": "MSFT"
}

start_date = "2020-01-01"


"""simulate last_ts"""


#df = pd.DataFrame()
df = yf.download(tickers=list(tickers.values()), start="2020-01-01", end="2020-12-31", interval="1d")["Close"]

def get_last_timestamp(df):
    if df.empty:
        return None
    return df.index.max()

last_ts = get_last_timestamp(df)


"""fetch_incremental.py"""


def fetch_incremental(last_ts=None):
    df = yf.download(tickers=list(tickers.values()),
                     start=last_ts.strftime("%Y-%m-%d") if last_ts is not None else start_date,
                     interval="1d"
                     )["Close"]
    return df

df = fetch_incremental(last_ts)
print(df)
```
