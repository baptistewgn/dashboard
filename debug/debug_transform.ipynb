```python
"""simulate config.py"""

windows = {
    1: "1d",
    5: "5d",
    21: "21d"
}
```

```python
import pandas as pd
from data_gen import generate_timeseries_multi

df = generate_timeseries_multi()
```

```python
out = df.copy()
rets_1d = df.pct_change(1).add_suffix("_1d_ret")
rets_5d = df.pct_change(5).add_suffix("_5d_ret")
rets_21d = df.pct_change(21).add_suffix("_21d_ret")
out = pd.concat([out, rets_1d, rets_5d, rets_21d], axis=1)

print(out)
```

```python
out = df.copy()

rets = []

for w, name in windows.items():
    r = df.pct_change(w).add_suffix(f"_{name}_ret")
    rets.append(r)

out = pd.concat([out] + rets, axis=1)

print(out)
```

```python
z63 = (df - df.rolling(63).mean()) / df.rolling(63).std()
z63.columns = [f"{c}_z63" for c in df.columns]

z252 = (df - df.rolling(252).mean()) / df.rolling(252).std()
z252.columns = [f"{c}_z252" for c in df.columns]

out = pd.concat([out, z63, z252], axis=1)

print(out)
```

```python
"""version to debug"""

def compute_features(df):
    out = df.copy()

    # returns
    rets = df.pct_change([1,5,21])
    rets.columns = [f"{c}_ret_{w}d" for c, w in rets.columns]
    out = pd.concat([out, rets], axis=1)

    # zscore
    z63 = (df - df.rolling(63).mean()) / df.rolling(63).std()
    z63.columns = [f"{c}_z63" for c in df.columns]

    z252 = (df - df.rolling(252).mean()) / df.rolling(252).std()
    z252.columns = [f"{c}_z252" for c in df.columns]

    out = pd.concat([out, z63, z252], axis=1)

    return out
```

```python
"""updated version"""

def compute_features(df):
    out = df.copy()
    
    #returns
    rets = []
    
    for w, name in windows.items():
        r = df.pct_change(w).add_suffix(f"_{name}_ret")
        rets.append(r)
    
    out = pd.concat([out] + rets, axis=1)

    # zscore
    z63 = (df - df.rolling(63).mean()) / df.rolling(63).std()
    z63.columns = [f"{c}_z63" for c in df.columns]

    z252 = (df - df.rolling(252).mean()) / df.rolling(252).std()
    z252.columns = [f"{c}_z252" for c in df.columns]

    out = pd.concat([out, z63, z252], axis=1)

    return out

print(compute_features(df))
```
